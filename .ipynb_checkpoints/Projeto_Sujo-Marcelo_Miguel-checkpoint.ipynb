## Pergunta a ser respondida:
 A entrada de um país na União Europeia faz com que haja um aumento no numero de mortes em suas rodovias?
  

# Hipotese e Mecanismo:
Quando um páis entra na União Europeia, ele deve atender ao critério de livre circulação de pessoas dos países membros. Dessa forma, a minha expectativa é a de que, com o aumento do numero de turismo, por consequência da entrana na UE, o número de mortes no trafego aumente em conjunto.

#### Importando bibliotecas e dados

```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import seaborn as sns
```

```python
traffic_deaths=pd.read_excel("traffic_deaths_per_100000_people.xlsx")
international_tourism_arrival=pd.read_excel("st_int_arvl.xlsx")
roads_paved=pd.read_excel("roads_paved_percent_of_total_roads.xlsx")
circulacao_veic=pd.read_excel("cars_trucks_and_buses_per_1000_persons.xlsx")
```

```python
traffic_deaths=traffic_deaths.set_index("country").transpose()
international_tourism_arrival=international_tourism_arrival.set_index("country").transpose()
roads_paved=roads_paved.set_index("country").transpose()
circulacao_veic=circulacao_veic.set_index("country").transpose()
```

```python
print("Def com os plt.plot")
def plota_grafico_UE(dado):
    plt.plot(dado.loc[(dado.index > 1975) & (dado.index < 2016), "Denmark"],label="Dinamarca")
    plt.plot(dado.loc[(dado.index > 1975) & (dado.index < 2016), "Norway"], label="Noruega")
    plt.plot(dado.loc[(dado.index > 1975) & (dado.index < 2016), "Sweden"],label="Suecia")
    plt.plot(dado.loc[(dado.index > 1975) & (dado.index < 2016), "Austria"], label="Austria")
    plt.legend()
    return (' ')
```

```python
traffic_deaths_original=traffic_deaths.transpose()
## Listando os países
lista_paises = ["Portugal", "Spain", "France", "United Kingdom", "Italy", 
     "Greece", "Norway", "Sweden", "Finland", "Austria", "Germany"]
lista_paises_europa =["Denmark","Norway","Sweden","Austria"]
## calculando mediana dos países da europa em mortes no tráfego
traffic_deaths_europa = traffic_deaths_original.loc[
    lista_paises,
    (traffic_deaths_original.columns > 1975) & (traffic_deaths_original.columns < 2016)]
median_europa = traffic_deaths_europa.median()
traffic_deaths_europa = traffic_deaths_europa.transpose()
traffic_deaths_europa['median'] = median_europa
```

```python
#dividindo essa mediana pelos países analisados
mediana_traffic=traffic_deaths_original.loc[lista_paises_europa,
    (traffic_deaths_original.columns > 1975) & (traffic_deaths_original.columns < 2016)]
mediana_traffic=mediana_traffic.transpose()
for p in lista_paises_europa:
    mediana_traffic[p] = mediana_traffic[p] / traffic_deaths_europa["median"]
```

```python
#calculando mediana dos países da europa em relação ao turismo e dividindo 
turismo_original=international_tourism_arrival.transpose()
turismo_europa=turismo_original.loc[lista_paises,(turismo_original.columns > 1975) & (turismo_original.columns < 2016)]
median_europa_turismo = turismo_europa.median()

turismo_europa=turismo_original.loc[lista_paises_europa,(turismo_original.columns > 1975) & (turismo_original.columns < 2016)]
turismo_europa=turismo_europa.transpose()
turismo_europa["median"]= median_europa_turismo
turismo_europa.transpose()
for p in lista_paises_europa:
    turismo_europa[p] = turismo_europa[p] / turismo_europa['median']
```

```python
plt.plot(median_europa, label="Mediana da Europa")
plt.grid(True)
print (plota_grafico_UE(traffic_deaths))
plt.title("Mortes no tráfego(a cada 100.000)")
plt.ylabel("Número de mortes")
```

```python
mediana_traffic[lista_paises_europa].plot.line()
```

```python
plt.grid(True)
print (plota_grafico_UE(international_tourism_arrival))
plt.plot(median_europa_turismo, label="Mediana da Europa")
plt.title("Chegada de turistas nos países")
plt.ylabel("Número de turistas")
```

```python
turismo_europa[lista_paises_europa].plot.line()
```

```python
traffic_austria=traffic_deaths.loc[(traffic_deaths.index > 1995) & (traffic_deaths.index < 2014), "Austria"]
international_arrival=international_tourism_arrival.loc[(international_tourism_arrival.index > 1995) & (international_tourism_arrival.index < 2014), "Austria"]
traffic_austriaa=traffic_austria.rename(columns={'Austria':'Mortes'})
df2=pd.concat([traffic_austriaa,international_arrival], axis=1)
df2=df2.rename(columns={0:"Mortes","Austria":"Turismo"})
df2.plot.scatter(x='Turismo',y='Mortes', loglog = False);
sns.regplot(df2["Turismo"],df2["Mortes"], color=('#a29bec'))
plt.title("Austria")
plt.grid(True)
print('A correlação entre Turismo e mortes é entre os anos de 1980 à 2000 é de', df2['Mortes'].corr(df2['Turismo']))
```
